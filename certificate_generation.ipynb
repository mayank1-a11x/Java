```python
# 📦 Step 1: Install dependencies
!pip install yagmail openpyxl

# 📁 Step 2: Upload Excel and ZIP
from google.colab import files
import zipfile, os
import yagmail
from openpyxl import load_workbook
import traceback

print("📤 Upload your Excel file and final_certificates.zip...")
uploaded = files.upload()

# ✅ Safety check for ZIP file
zip_files = [f for f in uploaded if f.endswith('.zip')]
if not zip_files:
    raise FileNotFoundError("❌ No .zip file uploaded. Please upload final_certificates.zip.")
zip_file = zip_files[0]

# ✅ Safety check for Excel file
excel_files = [f for f in uploaded if f.endswith('.xlsx')]
if not excel_files:
    raise FileNotFoundError("❌ No Excel (.xlsx) file uploaded. Please upload the Excel sheet.")
excel_file = excel_files[0]

# 🗃️ Step 3: Extract the ZIP
cert_dir = "certificates"
os.makedirs(cert_dir, exist_ok=True)
with zipfile.ZipFile(zip_file, 'r') as zip_ref:
    zip_ref.extractall(cert_dir)

# 📊 Step 4: Load Excel for names/emails
wb = load_workbook(excel_file)
sheet = wb.active
headers = [cell.value for cell in sheet[1]]
name_col = headers.index("Name") + 1
email_col = headers.index("Email Address") + 1

# 📧 Step 5: Gmail Setup
EMAIL_SENDER = "mayankjha.ai23@bmsce.ac.in"  # Your Gmail
EMAIL_SUBJECT = "Your Certificate for FUTURE OPPORTUNITIES - May 2025"

print("🔐 Please log in with your Gmail App Password (not your regular password)...")
yag = yagmail.SMTP(EMAIL_SENDER)

# 📤 Step 6: Send Emails (with null checks + error tracing)
for row in sheet.iter_rows(min_row=2, values_only=True):
    name_cell = row[name_col - 1]
    email_cell = row[email_col - 1]

    if not name_cell or not email_cell:
        print(f"⚠️ Skipping row due to missing name or email: {row}")
        continue

    name = name_cell.strip()
    email = email_cell.strip()
    cert_name = f"{name.replace(' ', '_')}_final.pdf"
    cert_path = os.path.join(cert_dir, cert_name)

    if os.path.exists(cert_path):
        body = f"""Dear {name},

Thank you for participating in the “FUTURE OPPORTUNITIES” event organized by Augment AI in collaboration with SIEC Education.

Please find your certificate of participation attached to this email.

Best regards,
Team Augment AI
"""
        try:
            yag.send(to=email, subject=EMAIL_SUBJECT, contents=body, attachments=cert_path)
            print(f"✅ Sent to {name} at {email}")
        except Exception as e:
            print("❌ Error Trace:")
            traceback.print_exc()
            print(f"❌ Failed to send to {name} at {email} - {str(e)}")
    else:
        print(f"⚠️ Certificate not found for {name}")
```
